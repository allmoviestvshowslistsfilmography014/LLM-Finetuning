```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "2" # 2.
```

```python
test_prompts = [
    # Basic Interaction Prompts
    "Hello, how can you help me today?",
    "What's the weather like in New York City?",
    "Can you tell me a joke?",
    "Who is the current President of the United States?",
    
    # Technical Inquiry Prompts
    "What's the difference between Python 2 and Python 3?",
    "Can you explain the concept of neural networks?",
    "How do I troubleshoot a '404 Not Found' error on my website?",
    "Tell me about the latest advancements in AI research.",
    
    # Personalized Assistance Prompts
    "I'm planning a trip to Paris. Can you recommend some attractions?",
    "I'm looking for a recipe for homemade pizza. Can you help?",
    "Can you remind me to call my doctor tomorrow at 10 AM?",
    "What are some good exercises for strengthening my back?",
    
    # Ethical Inquiry Prompts
    "What are the ethical considerations in AI development?",
    "How do companies ensure data privacy?",
    "What's your opinion on autonomous vehicles?",
    "Can you describe the concept of fairness in machine learning?",
    
    # Emotion Handling Prompts
    "I'm feeling down today. Can you cheer me up?",
    "I'm so excited about my new job! What's your response?",
    "I'm really frustrated with my computer issues. Can you provide assistance?",
    "I'm curious about the universe. Can you tell me something fascinating?",
    
    # Entertainment and Pop Culture Prompts
    "Can you recommend a good sci-fi book?",
    
    # Business and Finance Prompts
    "How does the stock market work?",
    "Can you explain cryptocurrency?",
    "Tell me about Warren Buffett's investment strategy.",
    
    # Health and Lifestyle Prompts
    "What are some healthy eating habits?",
    "Can you describe a good fitness routine for beginners?",
    "Tell me about the benefits of mindfulness meditation.",
    "How can I improve my sleep quality?",
    
    # Educational Assistance Prompts
    "Can you help me understand Pythagoras' theorem?",
    "What's the capital of Australia?",
    "Describe the water cycle to me.",
    "Can you summarize the plot of 'To Kill a Mockingbird'?",
    
    # Travel and Exploration Prompts
    "What are some must-see places in Japan?",
    "Tell me about the Seven Wonders of the World.",
    "What's the best time to visit the Grand Canyon?",
    
    # Language and Communication Prompts
    "How do you say 'Thank you' in French?",
    "Can you explain the difference between 'their', 'there', and 'they're'?",
    "Tell me about the origins of the English language.",
    "What are some common idioms and their meanings?",
    
    # AI and Technology Trend Prompts
    "What's the latest development in Quantum Computing?",
    "Tell me about the Internet of Things (IoT).",
    "How is virtual reality used in education?",
    "What are the implications of 5G technology?"
]
```

```python
#https://huggingface.co/docs/peft/quicktour 

from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer
import torch

model = AutoPeftModelForCausalLM.from_pretrained("Sentdex/Walls1337bot-Llama2-7B-003.005.5000")
tokenizer = AutoTokenizer.from_pretrained("NousResearch/Llama-2-7b-chat-hf")

model = model.to("cuda")
model.eval()
```

```python
import colorama

for prompt in test_prompts:
    print(50*"=")
    formatted_prompt = f"### BEGIN CONVERSATION ###\n\n## Speaker_0: ##\n{prompt}\n\n## Walls1337bot: ##\n"
    inputs = tokenizer(formatted_prompt, return_tensors="pt")
    outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"), max_new_tokens=128)
    output_text = tokenizer.batch_decode(outputs.detach().cpu().numpy(), skip_special_tokens=True)[0]
    just_response = output_text.split(formatted_prompt)[1].split("\n\n##")[0]
    
    #print(formatted_prompt+just_response)
    print(f"{colorama.Fore.CYAN}{formatted_prompt}{colorama.Style.RESET_ALL}{colorama.Fore.YELLOW}{just_response}{colorama.Style.RESET_ALL}")
    print(50*"=")
    print()
    print()
```

```python
formatted_prompt = """### BEGIN CONVERSATION ###

## Speaker_0: ##
Can you tell me a joke?

## Walls1337bot: ##
Why did the chicken cross the playground?

## Speaker_0: ##
Idk, why?

## Speaker_1: ##
???
"""
inputs = tokenizer(formatted_prompt, return_tensors="pt")
outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"), max_new_tokens=128)
output_text = tokenizer.batch_decode(outputs.detach().cpu().numpy(), skip_special_tokens=True)[0]
just_response = output_text.split(formatted_prompt)[1].split("\n\n##")[0]

#print(formatted_prompt+just_response)
print(f"{colorama.Fore.CYAN}{formatted_prompt}{colorama.Style.RESET_ALL}{colorama.Fore.YELLOW}{just_response}{colorama.Style.RESET_ALL}")
```
